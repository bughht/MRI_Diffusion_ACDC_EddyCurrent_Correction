```python
import pypulseq as pp
import numpy as np
```

```python
system_low = pp.Opts(
    max_grad=30, grad_unit='mT/m',
    max_slew=30, slew_unit='T/m/s',
    rf_ringdown_time=20e-6, rf_dead_time=100e-6, adc_dead_time=20e-6
)

seq = pp.Sequence(system_low)
seq.add_block(pp.make_digital_output_pulse(channel='ext1', duration = 1e-3, system=system_low))
seq.add_block(pp.make_block_pulse(flip_angle=np.deg2rad(90), duration=3e-3, system=system_low))
seq.add_block(pp.make_delay(1e-3))
seq.add_block(pp.make_adc(num_samples=512, dwell=10e-6, system=system_low))
seq.write('seq/fieldmap_switch.seq')
```
